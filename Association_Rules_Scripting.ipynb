## Let's build BEST ASSOCIATES for UK as it has more transactions

```python
sa_data= online_retail_data[online_retail_data["Country"] == "Saudi Arabia"]
sa_data.head(20)
```

```python
uk_data= online_retail_data[online_retail_data["Country"] == "United Kingdom"]
uk_data.head(20)
```

```python
uk_data.isna().sum()
```

```python
1454/495478
```

```python
del uk_data["CustomerID"]
```

```python
uk_data.dropna(inplace = True)
```

```python
uk_data.isna().sum()
```

```python
#pd.pivot_table( data = sa_data,values="Quantity",index="InvoiceNo",columns="Description",aggfunc = "sum")
uk_pivotted_data = pd.pivot_table( data = uk_data,values="Quantity",index="InvoiceNo",columns="Description",aggfunc = "sum").fillna(0)
uk_pivotted_data
```

```python
uk_pivotted_data_onehot = uk_pivotted_data.applymap(lambda x:0 if x==0 else 1)
uk_pivotted_data_onehot
```

## 4. Data Mining

```python
from mlxtend.frequent_patterns import apriori,association_rules,
```

```python
freq_itemsets = apriori(df = uk_pivotted_data_onehot,min_support=0.02,use_colnames=True,)
freq_itemsets
```

```python
best_associates = association_rules(df = freq_itemsets ,metric='confidence',min_threshold=0.02,)
best_associates
```

```python
best_associates.to_csv("best_anticidents_consequents.csv")
```
